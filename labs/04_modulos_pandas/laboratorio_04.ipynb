## Problema 01

EL conjunto de datos se denomina `Automobile_data.csv`, el cual contiene información tal como: compañia, precio, kilometraje, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import numpy as np
import os
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","Automobile_data.csv")).set_index('index')
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Elimine los valores nulos (Nan)

```python
df=df.dropna() #reescribo mi dataframe sin las filas que tengan valores nulos
df.head()
```

2. Encuentra el nombre de la compañía de automóviles más cara

```python
grupo = df.groupby('company') #se agrupan datos por copañia
grupo_precio=grupo.agg({'price':[np.mean]}).reset_index() """se aplica el comando agg
para ver el promedio del precio de cada compañia"""
grupo_precio.max().to_frame() #se encuentra el maximo promedio de todas las compañias
```

3. Imprimir todos los detalles de Toyota Cars

```python
gt=df[df.company=='toyota']
gt.describe()
```

4. Cuente el total de automóviles por compañía

```python
compañias = df.groupby('company')  #se agrupa por compañia
compañias[['company']].count()#se cuentan los datos por compañia
```

5. Encuentra el coche con el precio más alto por compañía

```python
mask = (df.groupby(['company'])['price'].transform(max) == df['price'])
"""Por cada compañia, se compara el precio de cada dato con el maximo de la compañia"""
df[mask].reset_index()
#si se alcanza el maximo, se agrega el dato al dataframe
```

6. Encuentre el kilometraje promedio (**average-mileage**) de cada compañía automotriz

```python
grupo = df.groupby('company')
grupo_precio=grupo.agg({'average-mileage':[np.mean]}).reset_index()"""se aplica 
el comando agg para ver el promedio del average.milage de cada compañia"""
grupo_precio
```

7. Ordenar todos los autos por columna de precio (**price**)

```python
# Se ordenan todos los datos desde menor a mayor precio,
# muestro primeros 5 elementos solamente.
df.sort_values(['price']).reset_index().head()
```

## Problema 02


Siguiendo la temática de los automóviles, resuelva los siguientes problemas:

#### a) Subproblema 01
A partir de los siguientes diccionarios:

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'],
              'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 
                'Price': [29995, 23600, 61500 , 58900]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
* Concatene ambos dataframes (**carsDf**) y añada una llave ["Germany", "Japan"], según corresponda.

```python
carsDf1= pd.DataFrame(GermanCars)
carsDf2= pd.DataFrame(japaneseCars)
a=[carsDf1,carsDf2]
result=pd.concat(
    a, #ingreso mi arreglo de dataframes
    keys=['GermanCars','japaneseCars'], # defino mis nuevas Keys
    axis=1) # concateno de forma horizontal

result
```

#### b) Subproblema 02

A partir de los siguientes diccionarios:

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

* Cree dos dataframes (**carsDf1** y **carsDf2**) según corresponda.
*  Junte ambos dataframes (**carsDf**) por la llave **Company**.

```python
carsDf1=pd.DataFrame(Car_Price)
carsDf2=pd.DataFrame(car_Horsepower)
result = pd.merge(
    carsDf1,
    carsDf2 ,
    on='Company')
result
```
