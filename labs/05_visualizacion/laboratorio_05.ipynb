```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
palette = sns.color_palette("hls", 6)

sns.lineplot(
    x='month_number', #eje x
    y='total_profit', #eje y
    data=df,  #data frame
)   
```

```python
sns.scatterplot(
    x='month_number', #eje x
    y='total_profit',  # eje y
    data=df, # fuente de datos

)   
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python

fig = plt.figure(figsize=(10, 5)) 
colores=['b','g','r','c','m','y','b']

for i in range(1,7):
    plt.plot(
        df['month_number'],
        df[df.columns[i]],
        label=df.columns[i],
        color=colores[i],
        linewidth=1
)


plt.legend('venta de todos los productos') # agregar etiquetado
plt.title("gréfico multilinea") # agregar titulo
plt.xlabel('mes') # nombre eje x
plt.ylabel('cantidad') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python

fig = plt.figure(figsize=(10, 5)) #tamaño de ventana para los graficos
plt.subplot(1,2,1) #grafico de la izquierda
plt.bar(
    df['month_number'], # eje x
    df['facecream'], # eje y 
    align='center', # centrar nombre eje x
    color="g", # color 
    alpha=1 # intensidad del color
)
plt.subplot(1,2,2) #grafico de la derecha
plt.bar(
    df['month_number'], # eje x
    df['facewash'], # eje y 
    align='center', # centrar nombre eje x
    color="g", # color 
    alpha=1 # intensidad del color
)
plt.show()

```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Pre-format DataFrame
stats_df = df.drop(['month_number','total_units','total_profit'], axis=1)

 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
total={} #creo conjunto para almacenar las ventas totales de cada producto
for i in range(1,7): #agrego la suma total de las ventas a cada producto
    total[df.columns[i]]=df[df.columns[i]].sum() 
df_ventas_totales=pd.Series(total) #transformo el diccionario a serie
plt.pie(
    df_ventas_totales, #valores cuantitativos del grafico
    labels = df.drop(['month_number','total_profit','total_units'], axis=1).columns # sus respectivos nombres
)
plt.title('Ventas totales del año pasado',fontsize=18) #nombre del grafico
plt.show
```
